```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D

# Load the dataset
df = pd.read_excel("vis-data.xlsx")
df = df.drop(columns=['shearstress'])

# Set the color palette
palette = sns.color_palette("husl", 3)

# Line Plots
fig, ax = plt.subplots(3, 1, figsize=(10, 15))

sns.lineplot(x='nano', y='AV', data=df, ax=ax[0], color=palette[0])
ax[0].set_title('Nano vs AV')

sns.lineplot(x='T', y='AV', data=df, ax=ax[1], color=palette[1])
ax[1].set_title('Temperature vs AV')

sns.lineplot(x='shearrate', y='AV', data=df, ax=ax[2], color=palette[2])
ax[2].set_title('Shear Rate vs AV')

plt.tight_layout()
plt.show()
```

```python

# Bar Plots
fig, axs = plt.subplots(3, 1, figsize=(10, 15))

sns.barplot(x='nano', y='AV', data=df, ax=axs[0], palette=palette)
axs[0].set_title('Average AV by Nano Concentration')

sns.barplot(x='T', y='AV', data=df, ax=axs[1], palette=palette)
axs[1].set_title('Average AV by Temperature')

sns.barplot(x='shearrate', y='AV', data=df, ax=axs[2], palette=palette)
axs[2].set_title('Average AV by Shear Rate')

plt.tight_layout()
plt.show()
```

```python
palette = sns.color_palette("husl", 3)

# Joint Plots with different colors
sns.jointplot(x='nano', y='AV', data=df, kind='scatter', color=palette[0])
plt.show()

sns.jointplot(x='T', y='AV', data=df, kind='scatter', color=palette[1])
plt.show()

sns.jointplot(x='shearrate', y='AV', data=df, kind='scatter', color=palette[2])
plt.show()
```

```python
# Define the palettes
palette_husl = sns.color_palette("husl", 1)
palette_coolwarm = sns.color_palette("coolwarm", 2)

# lm Plots with different colors from different palettes
sns.lmplot(x='nano', y='AV', data=df, scatter_kws={'color': palette_husl[0]})
plt.title('Linear Relationship: Nano vs AV')
plt.show()

sns.lmplot(x='T', y='AV', data=df, scatter_kws={'color': palette_coolwarm[0]})
plt.title('Linear Relationship: Temperature vs AV')
plt.show()

sns.lmplot(x='shearrate', y='AV', data=df, scatter_kws={'color': palette_coolwarm[1]})
plt.title('Linear Relationship: Shear Rate vs AV')
plt.show()
```

```python

# Hexbin Plots with different colormaps
plt.hexbin(df['nano'], df['AV'], gridsize=30, cmap='Blues')
plt.colorbar()
plt.title('Hexbin Plot: Nano vs AV')
plt.xlabel('Nano')
plt.ylabel('AV')
plt.show()

plt.hexbin(df['T'], df['AV'], gridsize=30, cmap='Greens')
plt.colorbar()
plt.title('Hexbin Plot: Temperature vs AV')
plt.xlabel('Temperature')
plt.ylabel('AV')
plt.show()

plt.hexbin(df['shearrate'], df['AV'], gridsize=30, cmap='Purples')
plt.colorbar()
plt.title('Hexbin Plot: Shear Rate vs AV')
plt.xlabel('Shear Rate')
plt.ylabel('AV')
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 3D Scatter Plot 1 with 'plasma' colormap
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
sc = ax.scatter(df['nano'], df['T'], df['AV'], c=df['AV'], cmap='plasma', marker='o')
plt.colorbar(sc)
ax.set_xlabel('Nano')
ax.set_ylabel('T')
ax.set_zlabel('AV')
plt.title('3D Scatter Plot: Nano, Temperature, and AV (Plasma Colormap)')
plt.show()

# 3D Scatter Plot 2 with 'viridis' colormap
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
sc = ax.scatter(df['nano'], df['shearrate'], df['AV'], c=df['AV'], cmap='viridis', marker='o')
plt.colorbar(sc)
ax.set_xlabel('Nano')
ax.set_ylabel('ShearRate')
ax.set_zlabel('AV')
plt.title('3D Scatter Plot: Nano, Shear Rate, and AV (Viridis Colormap)')
plt.show()

# 3D Scatter Plot 3 with 'inferno' colormap
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
sc = ax.scatter(df['T'], df['shearrate'], df['AV'], c=df['AV'], cmap='inferno', marker='o')
plt.colorbar(sc)
ax.set_xlabel('Temperature')
ax.set_ylabel('ShearRate')
ax.set_zlabel('AV')
plt.title('3D Scatter Plot: Temperature, Shear Rate, and AV (Inferno Colormap)')
plt.show()
```
